На гистограмме распределния жилой площади мы можем увидеть, что больше всего домов имеют общую (жилую) площадь немного более 2000 квадратных метров. Также большой популярность пользуются дома до 2000 квадратных метров. Дома с большей жилой площадью встречаются реже.

Общая тенденция показывает, что строительство в этом районе сильно развивается, увеличая объемы с каждыом годом. Старые дома изживают себя и становятся непригодными для жизни.

**Условие 3:** 
3 задачаИсследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм 
из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.scatter(df['sqft_above'], df['price'])
plt.xlabel('Площадь')
plt.ylabel('Цена (в миллионах долларов)')
plt.title('Диаграмма рассеяния площади и цены')
plt.show()
```

Для исследования влияния площади домов на цену был выбран график рассеивания. На графике мы можем увидеть, что цена и площадь находятся в прямой пропорциональности, чем больше площадь, тем дороже недвижимость (что логично).

```python
plt.bar(df['bedrooms'], df['price'])
plt.xlabel('Количество комнат')
plt.ylabel('Цена (в миллионах долларов)')
plt.title('Гистограмма количества комнат и  цены')
plt.show()
```

На столбчатой диаграмме влияния количества спальных комнат на цену мы можем увидеть, что наиболее дорогие дома имеют 5-6 спален. 

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['floors'], y=df['price'])
plt.xlabel('Количество этажей')
plt.ylabel('Цена (в миллионах долларов)')
plt.title('Зависимость цены недвижимости от количества этажей')
plt.show()
```

На данном графике влияния количества этажей на стомость мы можем увидеть, что наиболее дорогими являются дома с двумя этажами и подвалом, так и квантили. и максимальное значение выше чем у остальных домов. 

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['grade'], y=df['price'])
plt.xlabel('Оценка')
plt.ylabel('Цена (в миллионах долларов)')
plt.title('Зависимость цены недвижимости от оценки')
plt.show()
```

На данном графике влияния состония  на стомость мы можем увидеть, что наиболее дорогими предсказуемо являются дома с высокой оценкой состояния, в целом, чем выше состояние дома, тем дороже продается дом.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['bathrooms'], df['price'], alpha=0.5)
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена (в миллионах долларов)')
plt.title('Зависимость цены недвижимости от количества ванных комнат')
plt.show()
```
